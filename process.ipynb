```python
import json
```

```python
with open("./data/task1_train.jsonl") as f:
    raw = f.readlines()
    data = []
    for line in raw:
        line = json.loads(line)
        data.append(line)
data[:1]
```

```python
task1 = []
max_length = 0
for line in data:
    for result in line['results']:
        length = 0
        pair = {}
        pair['input'] = line['context']
        length += len(line['context'])
        pair['output'] = str(result)
        length+=len(str(result))
        task1.append(pair)
        max_length = max(max_length,length)
task1
```

```python
print(max_length)
print(len(task1))
```

```python
with open("./data/task1_finetune.json", "w",encoding='utf-8') as f:
    data = json.dumps(task1, ensure_ascii=False,indent=4)
    f.write(data)
    
with open("./data/task1_finetune_small.json", "w",encoding='utf-8') as f:
    data = json.dumps(task1[:500], ensure_ascii=False,indent=4)
    f.write(data)
```
